## First model for inputting labels for unlabeled data

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import OneHotEncoder, PolynomialFeatures
from sklearn.model_selection import train_test_split, ParameterGrid
from sklearn.linear_model import LinearRegression, ElasticNet, Lasso
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.impute import SimpleImputer
#from sklearn_pandas import DataFrameMapper
from sklearn.metrics import mean_squared_error
from sklearn.compose import ColumnTransformer

import warnings
warnings.filterwarnings('ignore')
```

```python
modelling_df = labeled_df.drop(columns=['url', 'title', 'location', 'type', 'level', 'salaryRange',
                                'maxSalary', 'currency', 'salaryPeriod', 'hasSalary',
                                          'company', 'position', 'datePosted', 'validThrough', 'jobDescription', 
                                           'skills', 'lang', 'titleLang', 'details', 'finalJobDescription', 'scope'])
```

### Train-test split

```python
X = modelling_df.drop('minSalary', axis = 1)
y = modelling_df.minSalary
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
poly = ColumnTransformer([("poly", PolynomialFeatures(degree=2, interaction_only=True), slice(0, 296))], remainder='passthrough')
```

### Linear regression with interactions

```python
param_grid = dict(alpha=np.linspace(0.05, 1, 5).tolist(), 
                  l1_ratio=np.linspace(0, 1, 5).tolist())
param_grid
```

```python
results = []

for params in ParameterGrid(param_grid):
    print(params)
    
    pipe_reg = Pipeline(steps = [('poly', poly),
                                 ('en', ElasticNet(alpha=params['alpha'], l1_ratio=params['l1_ratio']))])
    
    pipe_reg.fit(X_train, y_train)
   
    results.append(dict(alpha=params['alpha'],
                        l1_ratio=params['l1_ratio'],
                        train_score=mean_squared_error(y_train,
                                                       pipe_reg.predict(X_train), 
                                                       squared=False),
        
                        test_score=mean_squared_error(y_test,
                                                      pipe_reg.predict(X_test), 
                                                      squared=False)
    ))
```

```python
pipe_reg = Pipeline(steps = [('poly', poly),
                                 ('en', ElasticNet(alpha=0.5, l1_ratio=0.5))])

pipe_reg.fit(X_train, y_train)
```

```python
rmse = mean_squared_error(y_test, pipe_reg.predict(X_test), squared=False)
rmse
```

```python
mape = mean_absolute_percentage_error(y_test, pipe_reg.predict(X_test))
mape
```

```python
res = pd.DataFrame(results)
res.sort_values('test_score').style.bar(subset=['train_score', 'test_score'], vmin=0)
```

### Decision Tree

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
param_grid = dict(n_estimators=[10, 50, 100, 200], 
                  max_depth=[10, 50, 100, 200],
                  min_samples_leaf = [1, 3, 5, 10], 
                  max_features = ["auto", "sqrt", "log2"])
param_grid
```

```python
from tempfile import mkdtemp
from shutil import rmtree
```

```python
cachedir = mkdtemp() # creates temp directory

results = []

for params in ParameterGrid(param_grid):
    print(params)
    
    pipe_reg = Pipeline(steps = [('poly', poly),
                                 ('rf_reg', RandomForestRegressor(
                                     n_estimators=params['n_estimators'], 
                                     max_depth=params['max_depth'],
                                     min_samples_leaf=params['min_samples_leaf'],
                                     max_features=params['max_features'],
                                     n_jobs = 2,
                                     random_state = 0,
                                     
                                     
                                 ))], memory=cachedir)
    
    pipe_reg.fit(X_train, y_train)
   
    results.append(dict(n_estimators=params['n_estimators'],
                        max_depth=params['max_depth'],
                        min_samples_leaf=params['min_samples_leaf'],
                        max_features=params['max_features'],
                        rmse_train_score=mean_squared_error(y_train,
                                                       pipe_reg.predict(X_train), 
                                                       squared=False),
        
                        rmse_test_score=mean_squared_error(y_test,
                                                      pipe_reg.predict(X_test), 
                                                      squared=False),
                        
                        mape_train_score=mean_absolute_percentage_error(y_train,
                                                       pipe_reg.predict(X_train)),
        
                        mape_test_score=mean_absolute_percentage_error(y_test,
                                                      pipe_reg.predict(X_test))
    ))
    
rmtree(cachedir) # removes temp directory
```

```python
pipe_reg = Pipeline(steps = [('poly', poly),
                                 ('rf_reg', RandomForestRegressor(
                                     n_estimators=200, 
                                     max_depth=200,
                                     min_samples_leaf=1,
                                     max_features='log2',
                                     n_jobs = 2,
                                     random_state = 0,
                                     
                                     
                                 ))])
```

```python
pipe_reg.fit(X_train, y_train)
```

```python
rmse = mean_squared_error(y_test, pipe_reg.predict(X_test), squared=False)
rmse
```

```python
mape = mean_absolute_percentage_error(y_test, pipe_reg.predict(X_test))
mape
```

```python
res = pd.DataFrame(results)
res.sort_values('rmse_test_score').style.bar(subset=['rmse_train_score', 'rmse_test_score', 'mape_train_score', 'mape_test_score'], vmin=0)
```

### KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
param_grid = dict(n_neighbors=[5, 10, 15], 
                  algorithm = ['ball_tree', 'kd_tree', 'brute'],
                  #leaf_size = [20, 30, 40],
                  metric = ['cosine', 'euclidean', 'minkowski', 'canberra', 
                            'chebyshev', 'braycurtis', 'cityblock']
                 )
param_grid
```

```python
from tempfile import mkdtemp
from shutil import rmtree
```

```python
cachedir = mkdtemp() # creates temp directory

results = []

for params in ParameterGrid(param_grid):
    print(params)
    
    try:
        pipe_reg = Pipeline(steps = [('poly', poly),
                                     ('knn_reg', KNeighborsRegressor(
                                         n_neighbors=params['n_neighbors'], 
                                         algorithm = params['algorithm'],
                                         metric = params['metric'],
                                         n_jobs = -1,
                                         weights = 'distance'


                                     ))], memory=cachedir)
    except ValueError as ve:
        continue
    
    pipe_reg.fit(X_train, y_train)
   
    results.append(dict(n_neighbors=params['n_neighbors'],
                        algorithm=params['algorithm'],
                        metric=params['metric'],
                        rmse_train_score=mean_squared_error(y_train,
                                                       pipe_reg.predict(X_train), 
                                                       squared=False),
        
                        rmse_test_score=mean_squared_error(y_test,
                                                      pipe_reg.predict(X_test), 
                                                      squared=False),
                        
                        mape_train_score=mean_absolute_percentage_error(y_train,
                                                       pipe_reg.predict(X_train)),
        
                        mape_test_score=mean_absolute_percentage_error(y_test,
                                                      pipe_reg.predict(X_test))
    ))
    
rmtree(cachedir) # removes temp directory
```
