# AI Jobs dataset (from index page, scraping all jobs with urls)

### 1. Libraries

```python
import csv
```

```python
import pandas as pd
```

```python
import time
from tqdm.auto import tqdm
```

```python
import selenium
from selenium import webdriver
from selenium.webdriver import FirefoxOptions
```

```python
opts = FirefoxOptions()
opts.add_argument('--headless')
browser = webdriver.Firefox(options=opts)
```

```python
from selenium.webdriver.common.by import By
```

```python
from selenium.common.exceptions import NoSuchElementException
```

### 2. Entering ai-jobs.net

```python
URL = 'https://ai-jobs.net'
browser.get(URL)
```

### 3. Open all vacancies through clicking "Load more" button

```python
import time
import random
```

```python
counter = 1
while counter < 38: 
    try:
        browser.find_element('link text', "Load more").click()
        print('click', counter)
        counter += 1
        time.sleep(random.randint(1, 3)+random.random())
    except Exception as e:
        print(e)
        break
```

```python
jobs = browser.find_elements(By.XPATH, "//a[@class='col list-group-item-action px-2 py-3']")
```

```python
len(jobs)
```

### 4. Scraping in chunks

```python
all_jobs3 = []
for job in tqdm(jobs[2000:]):
    time.sleep(.5)
    
    job_dict = {}
    job_dict['url'] = job.get_attribute(name='href')
    

    
    job_dict['title'] = job.get_attribute(name='title')
    

    
    try:
        job_dict['location'] = job.find_element(
            By.XPATH, ".//span[@class='d-none d-md-block text-break job-list-item-location']"
        ).text
    except NoSuchElementException:
        job_dict['location'] = None
    

    
    try:
        job_dict['type'] = job.find_element(
            By.XPATH, 
            ".//span[@class='badge badge-secondary badge-pill my-md-1']"
        ).text
    except NoSuchElementException:
        job_dict['type'] = None
    

    
    try:    
        job_dict['level'] = job.find_element(
            By.XPATH, 
            ".//span[@class='badge badge-info badge-pill my-md-1 d-none d-md-inline-block']").text
    except NoSuchElementException:
        job_dict['level'] = None
    

    
    try:
        job_dict['salary_range'] = job.find_element(
            By.XPATH, 
            ".//span[@class='badge badge-success badge-pill d-none d-md-inline-block']").text
    except NoSuchElementException:
        job_dict['salary_range'] = None
    

        
    try:
        job_dict['salary'] = job.find_element(
            By.XPATH, 
            ".//span[@class='badge badge-success badge-pill d-md-none']"
        ).text
    except NoSuchElementException:
        job_dict['salary'] = None
    

    
    try:
        job_dict['company'] = job.find_element(
            By.XPATH, 
            ".//p[@class='m-0 text-muted job-list-item-company']"
        ).text
    except NoSuchElementException:
        job_dict['company'] = None

    all_jobs3.append(job_dict)
```

### 5. Writing chunks to CSVs

```python
with open('aijobs3.csv','w') as f:
    w = csv.DictWriter(f, job_dict.keys())
    w.writerows(all_jobs3)
    f.close()
```

### 6. Merging all vacancies

```python
all_filenames = ['aijobs1.csv', 'aijobs2.csv', 'aijobs3.csv']
```

```python
aijobs1 = pd.read_csv("aijobs1.csv", header=None)
```

```python
aijobs2 = pd.read_csv("aijobs2.csv", header=None)
```

```python
aijobs3 = pd.read_csv("aijobs3.csv", header=None)
```

```python
concat = pd.concat([aijobs1, aijobs2, aijobs3], ignore_index=True)
```

```python
concat.rename(columns={0: 'url', 1: 'title', 2: 'location', 3: 'type', 4: 'level', 5: 'salaryRange', 6: 'salary', 7: 'company'}, inplace=True)
```
